## Dictionary

It's properties are:
* mutable
* uses key-value pairs to store data (keys must be unique and immutable)
* Dynamic, able to grow and shrink in size as needed
* uses the curly brackets (`{}`)
* accessing of dictionary elements is via the square brackets indexing operator (`[]`) and key.

General Syntax:
```python
d = {'key_1': 'value_1', ..., 'key_n': 'value_n'}
```

*Side Note:* the functions `keys()`, `values()` and `items()` return `view` objects which needs to be converted to be used.

## Sets

It's properties are:
* mutable
* stores **only** unique immutable objects
* unordered 

General Syntax:
```python
s = set(8,9,6,3,1)
```

2. Write function called `diff_dict()` that compares 2 dictionaries and returns a new dictionary of elements that contains the differences. Use the test cases to help you.

 **Program Output Example:**
<pre>
{}
{'c': [3, 4]}
{'c': [None, 4], 'd': [3, None]}
{'c': [3, None], 'd': [None, 4]}
</pre>

````python
def dict_diff(first, second):
    '''
    Function to find the different elements of a dictionary
    '''
    output = {}
    all_keys = set(first.keys()) | set(second.keys())

    for key in all_keys:
        if first.get(key) != second.get(key):
            output[key] = [first.get(key), second.get(key)]
    return output
````

## File Paths

Before any program is to start on file input and output, the program must be able to find the file (or determine its path). 

There are 2 types of file path terminology:
* **Absolute Path** - this refers to the **complete path details** to the location of the file or folder. 
* **Relative Path** - this refers to the a specific location of the file or folder **relative to** the current working directory. 

<br>

**What exactly is an Absolute Path and a Relative Path?**    
That would depend on the operating system that you are using. On a **Windows** machine, an absolute path is the full path (starts from the drive letter) of the file including the file name and extension and a relative path starts from either with a directory name or the file name with extension.

**Absolute Path**   
`c:\Windows\calc.exe` 

**Relative Path** (assuming that the executable is executed from the `Windows` directory)    
`calc.exe`

Python has 2 libraries that we can use to build file paths from your current working directory. The current working directory is the directory from which your python is executed from.
* **`os.path`** library -  this library is part of the `os` library that provides a portable way of using operating system dependent functionality. Uses the operating system operations build/manipulate file paths. Available in Python 2 and 3.
* **`pathlib`** library - Object oriented based modules that is able to build paths with the appropriate semantics for different operating systems. Use this if your program has no access to operating system operations. Python 3.4 and above.

## File Input/Output

Now that we have a rough understanding of file paths, let's learn about file input and output (File I/O for short). Regardless of the type of file that the program is trying to access, only 3 main steps are involved:

1. Open the file
2. Read or Write to the file
3. Close the file

The table below lists the functions that we can use to do those 3 steps mentioned above.

| Function | Description |
|:---|:---|
| `open()` | Open a file with the given name. |
| `close()` | Close the file. |
| `read()` | Read all data from the file. |
| `readline()` | Read one line at a time from the file. |
| `readlines()` | Read all lines from the file. |
| `write()` | Write data to the file. |
| `writelines()` | Write a list of lines to the file. |

### Opening and Closing a file

Before we can work with files, there's a few things that we need to answer first.
1. Does this file need to be created (aka this file is new)? 
2. If the file is an existing file, can we overwrite the data or append new data?
3. What is the file name? Must we built the file path, if yes where do we start from?
4. Which access mode are we using to open the file? Example of Access modes are `read`, `write`, `append`, etc.
5. Do we need to buffer any data?

**Important Note** ALWAYS make sure that your program has proper access rights to the files that it is going to access otherwise, you will need to ask your supervisor or IT department on how best to give your program those access rights.

**Let's answer these questions 1 at a time:**
1. Yes, the file is new so make sure to find out where this file can be created.


2. Ask your supervisor what to do with the existing data.


3. From the above section on *File Paths*, we roughly know how a path can be built via either the `os` or the `pathlib` library. Therefore we have ensured that program "can see" the file.


4. There are several access modes available and the table below lists the common ones. **Note** that access modes can be combined by constructing a string of desired access mode characters.

| Character | Meaning |
|:---|:---|
| `r` | Open for reading. **(default)** |
| `w` | Open for writing, overwriting all previous data in the file. Creates the file, if the file does not exists. |
| `a` | Open for writing and appending data to the end of the file (if the file already exists). |
| `x` | Open for exclusive creation meaning that if a file exists, a `FileExistsError` will be raised. Python 3.3 and above. |
| `b` | Binary mode. |
| `t` | Text mode. **(default)** Basically the file data is either in ASCII or UNICODE characters. |
| `+` | Open for updating (reading and writing). |
 
**Example** if we wanted to open a file for both reading and writing but without appending previous, we would use `r+` or `w+` because the `+` mode adds the opposing functionality to the first access mode.


5. Buffering of data is mainly used for data being received over a network or an older system where there is a latency in data reads. There are 3 buffering modes:
 * 0 - for no buffering
 * 1 - line buffering (only for text mode)
 * more than 1 - buffering action is performed with the indicated buffer size in bytes of a fixed-size chunk buffer. **Note** that negative number reverts back to system default.
 
 Depending on the way different organizations structure their file systems, buffering may or may not be required, again this is a question that you will need to ask your supervisor.

**Note** that getting the file creation or modification times are done via the `os` library and uses operating system specific function calls.

**Bonus Example: Cross-platform function to get the file's creation/modification date.**

````python
import os
import platform
import time

def creation_date(path_to_file):
    """
    Try to get the date that a file was created, falling back to when it was
    last modified if that isn't possible.
    Edited from question asked in http://stackoverflow.com/a/39501288/1709587.
    """
    if platform.system() == 'Windows':
        mod_time = time.ctime(os.path.getmtime(path_to_file))
        creat_time = time.ctime(os.path.getctime(path_to_file))
        
        return mod_time, creat_time
    else:
        stat = os.stat(path_to_file)
        try:
            return stat.st_birthtime
        except AttributeError:
            # We're probably on Linux. No easy way to get creation dates here,
            # so we'll settle for when its content was last modified.
            return stat.st_mtime


# print(creation_date(<some_file>))
````

As we can see from *example 3*, the `close()` function **always** follows after the `open()` function. This is a mandatory requirement because if the program terminates abruptly without closing a file, the file and its data will become  corrupted. To mitiagate this issue, Python has provided us an automatic mechanism for safer file open and close operations. 

This automatic mechanism is called a **Context Manager** and to use it, we use the `with` keyword. We will not be covering how to create our own *Context Managers* but we will be seeing more of this in the Database module.

**Example 4: Using a Context Manager to access a file.**

````python
with open("foo.txt", "r") as fo:
    print ("Name of the file: ", fo.name)
````

### Reading from and Writing to Files

Now that we have learnt how to open & close files and how to use a context manager, let's move on to reading and writing data to and from files.

Before we start, let's look at some terminology of a file:
* `EOF` - stands for `End of File`. Every file has an `EOF` character where it tells the computer's operating system that there is no more data to be read.
* `newline` - this differs between operating systems. In Unix based systems, it is the `\n` character but in the Windows systems it is the `\r\n` (Carriage Return & Newline). Depending on where and how the file was created, the `newline` characters will be different. We **will not** be using the Carriage Return character in this course but be aware that there are certain operating systems that uses it, [Wiki article](https://en.wikipedia.org/wiki/Newline#Representation).

**Reading from Files**   
From the table above, we have 3 functions that allows us to read data from files, let's investigate them.

1. `read(size=-1)` - this function reads and returns all the data from a file onto memory. **Do not** use this function for very large files unless you have the sufficient amount of RAM avaliable on your computer to store it. By default, the parameter `size` is `-1`, this means to read the file all the way to `EOF`. If a positive number is specified, it refers to the number of bytes to be read from the opened file.

 **Example 5: Reading all the data from the file using the `read()` function.**

````python
with open("foo.txt", "r") as fo:
    all_data = fo.read()

for data in all_data:
    print(data, end='')
````

2. `readline(size=-1)` this function reads and returns 1 line at a time from a file. A line is determined by the newline (`\n`)  character at the end of it. The `size` parameter has the same functionality as the `size` in the `read()` function. This function is normally coupled with a `while` loop to read all the lines in a file.

 **Example 6: Reading 1 line at a time from the file using `readline()` function.**

````python
with open("foo.txt", "r") as fo:
    while True:
        # reading 1 line at a time
        line = fo.readline()
        if not line:
            break
        print(line, end='')
````

3. `readlines(hint=-1)` this function reads and returns a list of lines from a file. Similar to the `readline()` function, the `hint=-1` parameter is optional and it refers to the number of bytes to be read from the lines in the opened file. If no size is given or a negative number, it reads the all the lines in the file.

 **Example 7: Read all lines in a file at once using the `readlines()` function.**

````python
with open("foo.txt", "r") as fo:
    all_lines = fo.readlines()

for line in all_lines:
    print(line, end='')
````

**Bonus**    
There is one more way that we can use to read data from a file without the use of any of the read functions. That is to use the file object itself, this works similarly to `readline()` function.

**Eamaple 8: Reading 1 line at a time using the file object as an iterator.**

````python
with open("foo.txt", "r") as fo:
    # the file object is an iterable object
    for line in fo:
        print(line, end='')
````

From *example 8*, we can deduce that the file object is an iterable object however, the `len()` function cannot be used on the it. If you would like to count the number of lines in a file, you would have to use variable as a counter while looping through all the lines in the file.

**Writing to Files**    
Now that we have learnt how to read data from a file, let's see how to write data to a file. Again, referring to the table from above, we can see that there are 2 functions for writing data to files.

**Note** that both write methods **do not** add the newline characters (`\n`) to the end of the string therefore it is up to us to format the string/s properly before writing it to a file.

1. `write()` - this function writes the data contained within a single string variable to the file.

 **Example 9: Writing data contained within a single string variable to a file.**

````python
str_val = 'We’ve all got both light and dark inside us.\n' + \
          'What matters is the part we choose to act on.\n' + \
          'That’s who we really are.'

# write to file
with open("foo.txt", "w") as fo:
    fo.write(str_val)
````

2. `writelines()` - this function writes a list of strings to the file. How is this different from the `write()` function? The difference is that `write()` function accepts single strings but `writelines()` accepts list of strings.

 **Example 10: Writing a list of strings to a file.**

````python
# create a list of strings
size = 6
seq = ["This is line number {}\n"] * size

# formating the strings
for i in range(size):
    seq[i] = seq[i].format(i+1)

# create then write to file
with open("foo.txt", "w") as fo:
    fo.writelines( seq )
    
print("File written successfully.")
print("File is located in the same directory as this notebook.")
````

Bear in mind that the string encoding matters when the data is written to file. **By default, all Python strings are using the `utf-8` character encoding.**

**Example 14: Plotting the the equation $ y=x^{5} - 8x^{3} + 10x + 6 $ with different line styles and colour.**

````python
import matplotlib.pyplot as plt
import decimal  # python's floating point arithmetic library
import math

# python has no range function for generating a range of uniform 
# floating point numbers therefore we need this function
def gen_floats(start, stop, step):
    '''
    Function to generate a range of uniform floating point numbers
    '''
    float_nums = []
    while start < stop:
        float_nums.append(start)
        start += decimal.Decimal(step)
    
    return float_nums
    
````

````python
# use a smaller step size for smoother curves 
x = gen_floats(-3, 3, 0.2)

# generating the corresponding y values
y = []
for num in x:
    y.append((num**5)-(8*(num**3))+(10*num)+6)

# plotting the line with style!
plt.plot(x, y, marker='o', linewidth=2, color='#F77538', linestyle='--')
plt.title("Plotting the equation $ y=x^{5} - 8x^{3} + 10x + 6 $")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.show()
````

Lets say that we want to plot 2 or more lines on to the same plot. For instance we want to check if there are any intersection points between these 2 math equations $ y=x^{5} - 8x^{3} + 10x + 6 $ and $ y=x^{4} - 8x^{2} $. Normally we would have to manually do some calculation to for points of intersection, point of inflection, etc but it would be easier to graph it and see whether or not we have to expand the effort.

**Example 15: Plotting 2 equations in the same plot.**

````python
# line 1 --------------------------------------------------
# use a smaller step size for smoother curves 
x = gen_floats(-3, 3, 0.2)

# generating the corresponding y values
y = []
for num in x:
    y.append((num**5)-(8*(num**3))+(10*num)+6)
    
# line 2 --------------------------------------------------
# use a smaller step size for smoother curves 
x1 = gen_floats(-4, 4, 0.2)

# generating the corresponding y values
y1 = []
for num in x1:
    y1.append((num**4)-(8*(num**2)))

# setting up the figure size
plt.figure(figsize=(10,8))

# plotting the line1
plt.plot(x, y, linewidth=2, color='#F77538', linestyle='--', 
         label='$ y=x^{5} - 8x^{3} + 10x + 6 $')
# plotting the line2
plt.plot(x1, y1, linewidth=2, color='#1AB75C', 
         label='$ y=x^{4} - 8x^{2} $')

# rest of the plot properties
plt.ylim(-20,20)
plt.grid(True)
plt.legend(bbox_to_anchor=(1,1))
plt.title("Two lines on the same graph")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.show()
````

If we do not want to plot 2 lines on the same plot, we can plot them in different `figure`s but within the same graph. This can be done via the function called `subplot()`. 

The function `subplot()` accepts either 3 comma separated integers (`3,2,1`) or a 3 digit integer (`321`). This integer tells the `subplot()` function how many rows and columns the plot is going have and index of the current plot being plotted. For example, the integer `311` means that there are 3 rows, 1 column and the current plot index is `1`. **Note that the `figure` indexes start from 1 and increases to the right.** Figure 4 below, shows a pictorial representation.

**Example 16: Plotting the math equations $ 5*sin(2x) $ and $ 10e^{-\frac{x}{3}}sin \pi x $ using `subplot()`**

````python
# line 1 --------------------------------------------------
# use a smaller step size for smoother curves 
x = gen_floats(0, 20, 0.1)

# generating the corresponding y values
y = []
for num in x:
    y.append(5*math.sin(2*num))
    
# line 2 --------------------------------------------------
# use a smaller step size for smoother curves 
x1 = gen_floats(0, 20, 0.1)

# generating the corresponding y values
y1 = []
for num in x1:
    y1.append(10 * math.exp(decimal.Decimal(-num/3)) * math.sin(decimal.Decimal(math.pi)*num))


# setting up the figure size
plt.figure(figsize=(10,8))

# plotting the lines in subplots
ax1 = plt.subplot(2,1,1)  # subplot with 2 rows, 1 column, index 1
plt.tight_layout()  # setting the plot layout
plt.subplots_adjust(hspace=0.2)  # adjusting the height of the space between subplots 
ax1.set_facecolor('#eafff5')

plt.plot(x, y, linewidth=2, color='#F77538')
plt.title("Harmonic Motion $ 5*sin(2x) $", fontsize='x-large')
plt.grid(True, axis='y')

# subplot 2
plt.subplot(2,1,2)  # subplot with 2 rows, 1 column, index 2
plt.plot(x1, y1, linewidth=2, color='#1AB75C')
plt.title(r"Damped Harmonic Motion $ 10e^{-\frac{x}{3}}sin \pi x $", fontsize='x-large')
plt.grid(True, axis='y')

plt.show()
````
